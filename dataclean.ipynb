```python
import pandas as pd
```

```python
data20 = pd.concat(map(pd.read_csv, ['202006-citibike-tripdata.csv', '202007-citibike-tripdata.csv', '202008-citibike-tripdata.csv']), ignore_index=True)
data20
```

```python
new = data20["starttime"].str.split(" ", n=1, expand = True)
data20["start date"]= new[0]
data20["start time"]= new[1]
data20.drop(columns =["starttime"], inplace = True)
data20
```

```python
new = data20["stoptime"].str.split(" ", n=1, expand = True)
data20["stop date"]= new[0]
data20["stop time"]= new[1]
data20.drop(columns =["stoptime"], inplace = True)
data20
```

```python
data20= data20[data20.gender !=0]
data20
```

```python
data20.to_csv("June-Aug 2020 data.csv", index=False, header=True)
```
